### Plotting in Matplotlib can be done in either of two ways, which ones? Which way is the recommended approach?

```python
Plotting i  Matplotlib kan göras på två sätt och det är objektorienterad metod samt state-machine metod. 
```

```python
import matplotlib.pyplot as plt

fig, ax = plt.subplots()  # Skapar figur och axlar
ax.plot([1, 2, 3], [10, 20, 30])  # Plottar på de specifika axlarna
ax.set_title('Objektorienterad plot')  # Lägger till titel
plt.show()
```

```python
import matplotlib.pyplot as plt

plt.plot([1, 2, 3], [10, 20, 30])  # Plottar direkt
plt.title('State-Machine plot')  # Lägger till titel
plt.show()
```

```python
Metoden som rekommenderas är objektorienterad metoden eftersom den har bättre kontroll och passar för större projekt. 
Medans State-Machine metoden används för enklare plottar.
```

### Explain shortly what a figure, axes, axis and an artist is in Matplotlib.

```python
En figur är den huvudsakliga behållaren för allt du ritar, exempelvis fönstret som visas.
Axes är grafen eller diagrammet i figuren, exempelvis grafen där linjen ritas.
En axis är en enskild axel i en axes, exempelvis x- och y-axeln i grafen.
En artist är allt som ritas på en axes, exempelvis linjen som ritas.
```

### When plotting in Matplotlib, what is the expected input data type?

### Create a plot of the function y = x^2 [from -4 to 4, hint use the np.linspace function] both in the object-oriented approach and the pyplot approach. Your plot should have a title and axis-labels.

```python
import numpy as np
import matplotlib.pyplot as plt

# Data: x-värden från -4 till 4 och y = x^2
x = np.linspace(-4, 4, 100)
y = x**2

# --- Objektorienterad metod ---
fig, ax = plt.subplots()
ax.plot(x, y)  # Plotta y = x^2
ax.set_title('y = x^2 (Objektorienterad)')  # Titel
ax.set_xlabel('x')  # x-axelns etikett
ax.set_ylabel('y')  # y-axelns etikett

# --- Pyplot-metod ---
plt.figure()
plt.plot(x, y)  # Plotta y = x^2
plt.title('y = x^2 (Pyplot)')  # Titel
plt.xlabel('x')  # x-axelns etikett
plt.ylabel('y')  # y-axelns etikett

# Visa båda graferna
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# --- Data for scatter plot ---
np.random.seed(15)
random_data_x = np.random.randn(1000)
random_data_y = np.random.randn(1000)
x = np.linspace(-2, 2, 100)
y = x**2

# --- Data for bar plot ---
fruit_data = {'grapes': 22, 'apple': 8, 'orange': 15, 'lemon': 20, 'lime': 25}
names = list(fruit_data.keys())
values = list(fruit_data.values())

# --- Create figure with subplots ---
fig, axes = plt.subplots(1, 2, figsize=(12, 5))

# Scatter plot in the first subplot
axes[0].scatter(random_data_x, random_data_y, alpha=0.6, label='Random Scatter')
axes[0].plot(x, y, color='red', label='y = x^2')  # Add curve for context
axes[0].set_title('Scatter Plot')
axes[0].set_xlabel('X-axis')
axes[0].set_ylabel('Y-axis')
axes[0].legend()
axes[0].grid(True)

# Bar plot in the second subplot
axes[1].bar(names, values, color='skyblue', alpha=0.8)
axes[1].set_title('Bar Plot')
axes[1].set_xlabel('Fruit')
axes[1].set_ylabel('Quantity')

# Adjust layout and show plot
plt.tight_layout()
plt.show()
```
